```python
def random_pont(canvas,N_points=400):
  # ランダムに1つのインデックスを選択
  ones_indices = np.where(canvas == 1)
  random_indices = np.random.choice(len(ones_indices[0]), N_points, replace=False)
  # 選択されたピクセルの座標を取得
  y_coords = ones_indices[0][random_indices]
  x_coords = ones_indices[1][random_indices]
  random_pixels = np.column_stack((y_coords, x_coords))
  return random_pixels
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
mask = np.ones((512,1024,1),np.uint8)
print(mask.shape)
points = random_pont(mask,10000)
print(points.shape)
plot = np.zeros((512,1024,1),np.uint8)

for i in points:
  plot[i[0],i[1],:] = 1
# print(canvas.shape)
plt.imshow(plot,cmap='gray')
cv2.imwrite('d_cam_all.png',mask)
# plt.imshow(mask,cmap='gray')
```
